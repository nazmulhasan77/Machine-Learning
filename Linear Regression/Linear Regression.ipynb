Linear Regression is a **supervised machine learning** algorithm used to **predict a continuous value** based on the relationship between one or more features (input variables) and a label (output variable). It assumes a **linear relationship** between them.

```python
import pandas as pd
```

```python
df=pd.read_csv('Salary Data.csv')
df.head() #Show first five row
```

```python
df.isnull().sum() #Show null value
```

```python
x = df.drop('Salary' ,axis=1) #Drop Level(output) keep only features
y = df['Salary'] #keep level into y 
```

```python
import matplotlib.pyplot as plt  
```

```python

plt.figure(figsize=(7,4) ) # canvas size
plt.scatter(x,y,label='Experiance & Salary Point') #scatter plot
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title('Experiance vs Salary')
plt.legend(loc='best',shadow=True) # show level , set location
plt.grid() # show grid
#plt.savefig('Experiance vs Salary Graph.png')
plt.show() # render view
```

```python
print(x.mean()) #finding mean value
print(y.mean()) #finding mean value

plt.figure(figsize=(7,4) ) # canvas size
plt.scatter(x.mean(),y.mean(),color="red",label='Mean Salary') #finding intersection point
plt.scatter(x,y,label='Experiance & Salary Point') #scatter plot
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title('Experiance vs Salary')
plt.legend(loc='best',shadow=True) # show level , set location
plt.grid() # show grid
#plt.savefig('Experiance vs Salary Graph.png')
plt.show() # render view
```

```python
import seaborn as sns #Plot Using Seaborn
sns.regplot(x=x,y=y,data=df) #Regression Plot
plt.savefig('Experiance vs Salary Regression Plot.png')
```

```python
from sklearn.linear_model import LinearRegression 
```

```python
# Train model
model = LinearRegression() #create object of LinearRegression() class
model.fit(x, y) #Training model using x and y
```

```python
m=model.coef_ 
c=model.intercept_
y_predicted=m*6+c
y_predicted
```

```python
# Predict
predicted = model.predict([[6]]) #Dataframe
print("Prediction for x=6 , y=", predicted)
```

```python
#Show all predection data
y_predicted = model.predict(x) #Dataframe
y_predicted
```

```python
df['y_predicted']=y_predicted #concatenate to df
df.head()
```

```python
plt.figure(figsize=(7,4) ) # canvas size
plt.scatter(x.mean(),y.mean(),color="red",label='Mean Salary') #finding intersection point
plt.scatter(x, y, color='blue', label='Actual Price')
plt.plot(x, df['y_predicted'], color='green', linewidth=2, label='Best Fit Line')
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title('Experiance vs Salary')
plt.legend(loc='best',shadow=True) # show level , set location
plt.grid() # show grid
plt.savefig('Experiance & Salary Actual vs Best Fit Line.png')
plt.show() # render view
```
